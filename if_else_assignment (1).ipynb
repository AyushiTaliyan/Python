#ques1: Q1. Write a program to calculate the electricity bill
using only if statement? (accept number of unit from user)

```python
num_of_units = int(input("Enter number of units: "))
```

```python
if(num_of_units <= 100):
    print ("No charge")
elif(num_of_units <=200):
    charge1 = (num_of_units - 100)*5
    print(charge1)
elif(num_of_units >200):
    charge2= 100*5 + (num_of_units -200)*10
    print(charge2)
else:
    print("NA")
```

```python
if(years_of_service > 10):
    net_bonus_amount = salary*(10/100)
    print(f"Net bonus amount is {net_bonus_amount}")
elif(years_of_service >= 6 and years_of_service <= 10):
    net_bonus_amount = salary*(8/100)
    print(f"Net bonus amount is {net_bonus_amount}")
elif(years_of_service < 6):
    net_bonus_amount = salary*(5/100)
    print(f"Net bonus amount is {net_bonus_amount}")
else:
    print("NA")
```

#Q6. Accept the marked price from the user and calculate
the Net amount as (Marked Price – Discount) to pay

```python
# ques6: find net amount
marked_price = int (input("Enter the marked price: "))
```

```python
if(marked_price > 10000):
    net_amount = marked_price - (marked_price*(20/100))
    print(f"Net Amount: {net_amount}")
elif(marked_price > 7000 and marked_price <= 10000):
    net_amount = marked_price - (marked_price*(15/100))
    print(f"Net Amount: {net_amount}")
elif(marked_price <= 7000):
    net_amount = marked_price - (marked_price*(10/100))
    print(f"Net Amount: {net_amount}")
else:
    print("NA")
```

#Q9. Write a program to check whether the last digit of a number
(entered by user) is divisible by 3 or not?

```python
#ques9: last digit divisible by 3 or not
num = int(input("Enter a number: "))
last_digit = num % 10
if(last_digit% 3 == 0):
    print(f"Last digit of a {num} is divisible by 3.")
else:
    print(f"Last digit of a {num} is not divisible by 3.")
```
